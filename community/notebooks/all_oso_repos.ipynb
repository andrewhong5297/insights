```python
import os
import pandas as pd
import sys

# import local dependencies
sys.path.append(os.path.abspath("../../scripts/"))
from oso_db import execute_query
```

```python
# gather all github artifacts
project_data = execute_query("""
    SELECT DISTINCT
        p.slug AS slug,
        p.name AS name,
        a.name AS repo
    FROM 
        project p
        INNER JOIN project_artifacts_artifact paa ON p.id = paa."projectId"
        INNER JOIN artifact a ON paa."artifactId" = a.id
    WHERE         
        a.namespace = 'GITHUB'
    ORDER BY 
        p.slug;
""", col_names=True)
```

```python
# dump the data to dsv
df = pd.DataFrame(project_data[1:], columns=project_data[0])
df
```
